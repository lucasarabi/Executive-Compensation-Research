## Prepare OriginalData.csv for reformat

```python
# Read original data import

original_data_import = pd.read_csv('../../data/OriginalData.csv')

display(original_data_import.head())

dataframe = original_data_import

# 'original_data_import' is intended to be READY-ONLY
# From this point on, morph dataset using 'dataframe'
```

```python
# Index by Date column, drop Date column as feature 

# First, cast Dates to datetime format
dataframe['Dates'] = pd.to_datetime(dataframe['Dates'])

date_index_dataframe = dataframe.set_index('Dates')

# 'date_index_dataframe' now indexes by what was originally the Date column
date_index_dataframe.head()
```

```python
# Cast other dates to datetime format

# Find columns in need of cast
date_index_dataframe.info()
# 1. DATE_OF_LAST_EXECUTIVE_CHANGE
# 2. DATE_OF_LAST_BOD_CHANGE 
 
date_index_dataframe['DATE_OF_LAST_EXECUTIVE_CHANGE'] = pd.to_datetime(date_index_dataframe['DATE_OF_LAST_EXECUTIVE_CHANGE'])
date_index_dataframe['DATE_OF_LAST_BOD_CHANGE'] = pd.to_datetime(date_index_dataframe['DATE_OF_LAST_BOD_CHANGE'])

# Compare to original .info() call for 15748 non-null count for datetime64[ns] on denoted columns
print('\n\n')
date_index_dataframe.info()
```
