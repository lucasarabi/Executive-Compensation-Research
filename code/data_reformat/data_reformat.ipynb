```python
# Libraries

import os 
import pandas as pd
import numpy as np
```

## Prepare OriginalData.csv for reformat

```python
# Read original data import

original_data_import = pd.read_csv('../../data/OriginalData.csv')

display(original_data_import.head())

dataframe = original_data_import

# 'original_data_import' is intended to be READY-ONLY
# From this point on, morph dataset using 'dataframe'
```

```python
# Index by Date column, drop Date column as feature 

# First, cast Dates to datetime format
dataframe['Dates'] = pd.to_datetime(dataframe['Dates'])

date_index_dataframe = dataframe.set_index('Dates')

# 'date_index_dataframe' now indexes by what was originally the Date column
date_index_dataframe.head()
```

```python
# Cast other dates to datetime format

# Find columns in need of cast
date_index_dataframe.info()
# 1. DATE_OF_LAST_EXECUTIVE_CHANGE
# 2. DATE_OF_LAST_BOD_CHANGE 
 
date_index_dataframe['DATE_OF_LAST_EXECUTIVE_CHANGE'] = pd.to_datetime(date_index_dataframe['DATE_OF_LAST_EXECUTIVE_CHANGE'])
date_index_dataframe['DATE_OF_LAST_BOD_CHANGE'] = pd.to_datetime(date_index_dataframe['DATE_OF_LAST_BOD_CHANGE'])

# Compare to original .info() call for 15748 non-null count for datetime64[ns] on denoted columns
print('\n\n')
date_index_dataframe.info()
```

DO NOT RUN THIS SECTION MORE THAN ONCE

```python
# One hot encode categorical data; see last .info() call to find desired columns
# 1. symbol
# 2. gics_sector_name
# 3. gics_sub_industry_name
# 4. gics_industry_name
# 5. gics_industry_group_name
# 
# ** gics_industy and gics_sector are already int64
#
# One hot encoding will be done with natively with pandas

# Initial .info() for comparison to final output
date_index_dataframe.info()

# columns to one hot encode
# EXCLUDING 'symbol' COLUMN; THE DUMMY VARIABLE DATAFRAME IS BEING INITIALIZED WITH IT 
columns = ['gics_sector_name', 'gics_sub_industry_name', 'gics_industry_name', 'gics_industry_group_name']
dummy_variables = pd.get_dummies(date_index_dataframe['symbol'])

for i in columns:
    temp = pd.get_dummies(date_index_dataframe[i])
    dummy_variables = pd.concat([dummy_variables, temp], axis=1)

print('Final dummy variable dataframe to be joined')
display(dummy_variables)
print('\n\n')
print('Check dimensions (row number) against that of dummy variable dataframe')
display(date_index_dataframe)
print('\n\n')

# Concatenate dummy variables to left side of dataframe
ohe_dataframe = pd.concat([dummy_variables, date_index_dataframe], axis=1)
# Drop one hot encoded columns
# symbol, columns list
ohe_dataframe = ohe_dataframe.drop('symbol', axis=1)
ohe_dataframe = ohe_dataframe.drop(columns, axis=1) 

# Display dataframe, check dimensions
print('One hot encoded dataframe')
display(ohe_dataframe.head())
ohe_dataframe.info()

# Assure there are no objects in .info() call
```
